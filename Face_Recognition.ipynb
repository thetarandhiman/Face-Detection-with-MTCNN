```python
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle

def draw_patches(filename, result_list):

  data = pyplot.imread(filename)
  pyplot.imshow(data)
  ax = pyplot.gca()

  for result in result_list:
    x, y, width, height = result['box']

    rect = Rectangle((x,y), width, height, fill= False, color = 'red')

    ax.add_patch(rect)

    for key, value in result['keypoints'].items():
      dot = Circle(value, radius=2, color='blue')
      ax.add_patch(dot)


pyplot.show()
```

```python
def draw_faces(filename, result_list):
  data = pyplot.imread(filename)

  for i in range(len(result_list)):
    x1, y1, width, height = result_list[i]['box']
    x2, y2 = x1 + width, y1 + height

    pyplot.subplot(1, len(result_list), i+1)
    pyplot.axis('off')

    pyplot.imshow(data[y1:y2, x1:x2])

pyplot.show()
```

```python
test1 = 'test1.jpg'
pixels = pyplot.imread(test1)
detector = MTCNN()
faces1 = detector.detect_faces(pixels)
for face in faces1:
  print(face)
```

```python
draw_patches(test1, faces1)
```

```python
draw_faces(test1, faces1)
```

```python
test2 = 'test2.jpg'
pixels = pyplot.imread(test2)
detector = MTCNN()
faces2 = detector.detect_faces(pixels)
for face in faces2:
  print(face)
```

```python
draw_patches(test2, faces2)
```

```python
draw_faces(test2, faces2)
```
